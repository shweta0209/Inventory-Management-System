```python
File=open("Products.json","r")
prod=File.read()
products=json.loads(prod)
File.close()
```

```python
import time
ch="yes"
Total_bill=0

while ch == "yes":
    p_id = input("Enter the product id : ")
    if (int(p_id)<3001 or int(p_id)>3040):
        print("invalid Id")
        p_id=input("Enter the another product id : ")
        
    quantity=int(input("Enter the how much Quantity : "))

    if(quantity > products[p_id]['Quantity']):
        print("only {} products available.\n".format(products[p_id]['Quantity']))
    else:
        print("________________________________________________")
        print("Product id : ",p_id)
        print("product name : ", products[p_id]['Name'])
        print("Available Quantity : ", products[p_id]['Quantity'])
        print("Price per piece : ",products[p_id]['MRP'])
        print("Manufacturing Date : ",products[p_id]['MFD Date'])
        print("Expiry Date : ",products[p_id]['Expiry Date'])
        print("------------------------------------------------")
        print("final price :", (products[p_id]['MRP'])*quantity)
        Total_bill+=(products[p_id]['MRP'])*quantity
        
        products[p_id]['Quantity']=products[p_id]['Quantity']-quantity
    
    #updating sales file and adding timestamp
    Sale=open("Sales.json","a")
    sale=json.dumps(products[p_id])
    Sale.write(sale+time.ctime()+"\n")
    Sale.close()
    
    
    ch=input("Enter 'yes' for continue buying, else enter no : ")
print("________________________________________________")   
print("TOTAL AMOUNT : ", Total_bill)
print("________________________________________________")
```
